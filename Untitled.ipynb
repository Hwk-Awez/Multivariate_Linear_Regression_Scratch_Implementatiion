```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('WineQT.csv')
df.head()
```

```python
df.drop(['Id'],axis=1,inplace=True)
```

```python
df.columns
```

```python
x = df.drop(['quality'],axis=1)
y = df['quality']
```

```python
print(x.shape)
print(y.shape)
```

```python
x=np.array(x)
y=np.array(y)

print(x)
print(y)
```

```python
m=len(x)
m
```

```python
theta = np.zeros(len(x[0]))
c = np.zeros(len(x[0]))
print(theta)
print(c)
```

```python
def cost_func(x,y,theta,c,m):

    sum = 0

    for i in range(m):
        temp = 0
        temp += np.dot(theta,x[i])
        temp -= y[i]
        temp += np.sum(c)
        sum += temp ** 2;

    return (sum / (2*m))
```

```python
cost_func(x,y,theta,c,m)
```

```python
alpha = 0.0001
epochs = 1000
```

```python
grad_mse=[]
grad_theta=[[] for _ in range(len(theta))]

for i in range(epochs):
    
    for j in range(len(theta)):
        sum =0

        for k in range(len(x)):

            temp = 0

            temp += np.dot(theta,x[k])
            temp += np.sum(c)
            temp -= y[k]

            sum += temp

        sum /= m

        theta[j] = theta[j] - (alpha * sum)

        grad_theta[j].append(theta[j])

    
    grad_mse.append(cost_func(x,y,theta,c,m))
    
    
    if((i+1) % 100 ==0):
        print(cost_func(x,y,theta,c,m))
```

```python
theta
```

```python
print("Mean error:/n")

error= 0

for i in range(m):

    error += np.dot(theta,x[i]) - y[i]

error /= m

print(error)
```
